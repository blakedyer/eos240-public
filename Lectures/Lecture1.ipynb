```python
Mg = 24.3
Si = 28.1
Ca = 40.1
Fe = 55.8
Al = 27
O = 16

MgO = Mg + O
SiO2 = Si + 2 * O
Al2O3 = Al * 2 + O * 3
CaO = Ca + O
FeO = Fe + O

(104 * MgO) / (104 * MgO + 100 * SiO2 + 0.5 * 8.4 * Al2O3 + 6.2 * CaO + 86 * FeO) * 100
```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(10,10))
frac = np.linspace(0, 1, 100)
total_M = 104 * MgO + 100 * SiO2 + 0.5 * 8.4 * Al2O3 + 6.2 * CaO + (1 - frac) * 86 * FeO

plt.plot(frac, (104 * MgO) / total_M * 100, label="MgO")
plt.plot(frac, (1 - frac) * 86 * FeO / total_M * 100, label="FeO")
plt.gca().set_xlabel("Fraction of FeO removed to Core")
plt.gca().set_ylabel("Wt % Oxide")
plt.legend()
```
